```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_eval_directa = pd.read_csv(
    "../data/Eval_directa.csv", 
    sep=';'
    )

display(df_eval_directa)

plt.figure(figsize=(9, 5.5))

plt.plot(
    df_eval_directa['Tamaño (n)'],
    df_eval_directa[' Timepo (Microsegundos)'],
    marker = 'o',
    color = '#1f77b4',
    linestyle = '-',
    linewidth = 2,
    label = 'Algoritmo 1: Evaluación directa $O(n²)$'
)

plt.title('Tarea Unidad III: Tiempo de ejecución en función de $n$')
plt.xlabel('Tamaño del polinomio ($n$)', fontsize = 11)
plt.ylabel('Tiempo de ejecución (Microsegundos)', fontsize = 11)

plt.xticks(df_eval_directa['Tamaño (n)'])

plt.grid(True, linestyle = '--', alpha = 0.6)

plt.legend(fontsize = 11, loc = 'upper left')

plt.tight_layout()
plt.show
```
